```python
import pandas as pd
```

```python
Df=pd.read_csv('Sample 2ND YR.xls')
Df.head()
```

```python
data=Df.iloc[:,[0]+list(range(56,107))]
data
```

```python
newDf=data.copy()
newDf.drop(['CANDIDATE ID','TOTR'],axis=1,inplace=True)
newDf
```

```python
newDf.shape[0]
```

```python
marks=[]
wrong=[]
right=[]
n=0
while(n<newDf.shape[0]):
    m=0
    r=0
    w=0
    for i in newDf:
        if newDf[i][n]==newDf[i][0] or i=='Q70':
            m+=1
            r+=1
        else:
            w+=1
    marks.append(m)
    right.append(r)
    wrong.append(w)
    n+=1
```

```python
data['DA Marks']=marks
data['Right DA']=right
data['Wrong']=wrong
data
```

```python
data.sort_values(by='DA Marks',ascending=False, inplace=True)
data
```

```python
data['Rank']=list(range(0,data.shape[0]))
data
```

```python
rankDf=data[['Rank','CANDIDATE ID','DA Marks','TOTR']]
rankDf['DSA Marks']=rankDf['TOTR']-rankDf['DA Marks']
rankDf
```
